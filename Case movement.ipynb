```python
import pandas as pd
```

```python
df = pd.read_csv("data/case-movement.csv")
df.head()
```

```python
def convert_year(yr):
    last_two = int(yr[-2:])
    if last_two > 90:
        return 1900 + last_two
    return 2000 + last_two
    
    return last_two

convert_year("1994-95")
```

```python
df["Year"] = df["Fiscal Year"].apply(convert_year)
df
```

```python
def fix(badn):
    return int(str(badn).replace(",",""))

for c in df.columns[1:]:
    df[c] = df[c].apply(fix)

df.dtypes
```

```python
import matplotlib
#matplotlib.use( "agg" )
# matplotlib.use( "macosx" )

%matplotlib inline
# matplotlib.style.use('ggplot')
matplotlib.style.use('mirror00')
import matplotlib.pyplot as plt
```

```python
import os 
def plot(offdf,fname, title="",subfolder="",figsize=(3,3),ylabel="",xlabel=""):
    tmpdf = offdf.copy()
    #fname = tmpdf["Title"][tmpdf.index[0]]
    ax = tmpdf.plot(figsize=figsize)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    df.index = df.index.astype(int)
    #ax.set_title(off_df["title"].iloc[0])
    ax.set_title(title)
        
    plt.tight_layout()
    
    plt.savefig(os.path.join("img", subfolder, fname + ".png"),
                bbox_inches="tight")
    plt.close()


plot(df.set_index("Year")["JD Disposed"],"jd-disposed",figsize=(6,4),subfolder="movement")
```

```python
df["pct"] = df["JD Disposed"] * 100 / df["JD Disposed"].max()
plot(df.set_index("Year")["pct"],"jd-disposed",figsize=(6,4),subfolder="movement")
```

```python
firearms = pd.read_csv("data/firearm-only.csv").set_index("Year")
firearms["pct"] = firearms["Charges"] * 100 / firearms["Charges"].max()
firearms
```

```python
plot(firearms[firearms.index >= 2002]["pct"],
     "jd-firearm-combined",figsize=(6,4),subfolder="movement")
     
```

```python
combined = firearms.join(df.set_index("Year"),how="left",lsuffix="_Firearm",rsuffix="_JD")
combined = combined[["pct_Firearm","pct_JD"]]
combined.columns = ("Firearms charges","All cases")
```

```python
plot(combined,"combined",figsize=(6,4),subfolder="movement",
     xlabel="Fiscal year",ylabel="Percent of peak")
```
